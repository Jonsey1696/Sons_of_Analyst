```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# for now, just use the dictionary to test the code
src_words = pd.read_csv('skills.csv')
```

```python
# initialize string
cloud_text = ''

# build a giant string with all words
for word in src_words['Job Skill']:
     cloud_text = cloud_text + ' ' + str(word)

# define WordCloud
cloud = WordCloud(width = 1200, height = 800, background_color='white').generate(text=cloud_text)

# plot the figure and show
plt.figure(figsize = (12, 8), facecolor = None) 
plt.imshow(cloud)
plt.axis('off')
#plt.tight_layout(pad = 0)
plt.show()

```
