```python
import pandas as pd
```

```python
csv_path="../Sons_of_Analyst/data_scientist_united_states_job_postings_jobspikr.csv"

kaggle_data_df=pd.read_csv(csv_path)
```

```python
kaggle_data_df["category"].value_counts()
```

```python
kaggle_data_df["job_title"].tolist()
```

```python
kaggle_data_df["job_title"] = kaggle_data_df.job_title.str.lower()
kaggle_data_df
```

```python
groups=kaggle_data_df.groupby(["job_title"]).count()
```

```python
jrs=kaggle_data_df[kaggle_data_df["job_title"].str.contains("jr|junior|entry|associate")]
jrs["job_title"]="junior data scientist"
jrs
```

```python
seniors=kaggle_data_df[kaggle_data_df["job_title"].str.contains("senior|lead|principal")]
seniors["job_title"]="senior data scientist"
seniors
```

```python
frames=[jrs, seniors]
```

```python

jrs_seniors=pd.concat(frames)
```

```python
jrs_seniors
```

```python
jrs_seniors.reset_index(inplace=True)
jrs_seniors
```

```python
drop_jrs_seniors=jrs_seniors["index"].tolist()
```

```python
drop_jrs_seniors
```

```python
just_data=kaggle_data_df.drop(drop_jrs_seniors)
```

```python
just_data["job_title"]="data scientist"
```

```python
frames=[just_data, jrs_seniors]
clean_roles=pd.concat(frames)
clean_roles
```

```python
duplicates=kaggle_data_df.drop_duplicates(subset=["url"])
duplicates
```

```python
clean_roles2=clean_roles.drop_duplicates(subset=["url"])
clean_roles2["job_title"].value_counts()
```

```python
clean_roles2.to_csv("clean_data.csv")
```

```python
clean_roles2=clean_roles2[[
 'job_title',
 'category',
 'company_name',
    'url',
 'city',
 'state',
 'country',
 'job_description',
 'job_type',
 'salary_offered',
 'job_board']]
```

```python
clean_roles2.to_csv("clean_data.csv")
```
