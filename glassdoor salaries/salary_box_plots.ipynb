```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import master df

jobs_df=pd.read_csv("Coordinates.csv")
jobs_df.drop(columns="Unnamed: 0")
jobs_df = jobs_df[jobs_df["salary_average"] != 0]
```

```python
#Create subset dfs

junior_df=jobs_df[jobs_df["Job Title"]=="junior data scientist"]
mid_level_df=jobs_df[jobs_df["Job Title"]=="data scientist"]
senior_df=jobs_df[jobs_df["Job Title"]=="senior data scientist"]
```

```python
junior_salaries=junior_df["salary_average"].tolist()
mid_salaries=mid_level_df["salary_average"].tolist()
senior_salaries=senior_df["salary_average"].tolist()
```

```python
jr_count = junior_df["salary_average"].count()
m_count = mid_level_df["salary_average"].count()
sr_count = senior_df["salary_average"].count()
jr_count, m_count, sr_count
```

```python
import statistics as sts
```

```python
senior_salaries.sort()
print(sts.mean(senior_salaries))
```

```python
salaries = [junior_salaries, mid_salaries, senior_salaries]
levels=["Junior", "Mid-Level", "Senior"]
green_marker = dict(markerfacecolor='g', marker='o')

fig1, ax1 = plt.subplots()
ax1.set_title("Average Salary by Role Level")
ax1.boxplot(salaries, flierprops=green_marker)
ax1.set_ylabel('Salary')
plt.xticks([1, 2, 3], levels)

plt.savefig("average_salary_whiskers.png")
```
