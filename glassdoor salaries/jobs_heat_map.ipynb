```python
import gmaps
import pandas as pd
from matplotlib import pyplot as plt

from config import gkey

gmaps.configure(api_key=gkey)
```

```python
jobs_df=pd.read_csv("Coordinates.csv")
jobs_df.head()
```

```python
locations=jobs_df[["Lat", "Lng"]]
salary=jobs_df["salary_average"]
```

```python
jobs_df["salary_average"].max()
```

```python
fig=gmaps.figure()

heat_layer=gmaps.heatmap_layer(locations, weights=salary, 
                                 dissipating=False, max_intensity=184000,
                                 point_radius=1)
fig.add_layer(heat_layer)

fig
```

```python
plt.savefig("All_jobs_heatmap.png")
```
